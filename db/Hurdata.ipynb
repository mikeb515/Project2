```python
import pandas as pd
import numpy as np
import flask
import json as js
import pyproj as pp
```

```python
pacific = pd.read_csv("pacific.csv")
atlantic = pd.read_csv('atlantic.csv')

print(len(pacific))
print(len(atlantic))
```

```python
df = [pacific, atlantic]
```

```python
hurdata = pd.concat(df)
```

```python
hurdata = hurdata[['ID','Name','Date','Time','Event','Status','Latitude','Longitude','Maximum Wind','Minimum Pressure']]
hurdata
```

```python
Lattitude = []
Longitude = []
for idx, row in hurdata.iterrows():
    Lat = row["Latitude"]
    Lon = row['Longitude']
    if 'N' in Lat:
        Lattransform = float(Lat.rstrip('N'))
    else: 
        Lattransform = -float(Lat.rstrip('S'))
        
    if "W" in Lon:
        Lontransform = -float(Lon.rstrip('W'))
    else: 
        Lontransform = float(Lon.rstrip('E'))
    Lattitude.append(Lattransform)
    Longitude.append(Lontransform)
    
    print((idx+1)/len(hurdata))
    
        
```

```python
hurdata['Latitude'] = Lattitude
hurdata['Longitude'] = Longitude
hurdata.head(100)
```
